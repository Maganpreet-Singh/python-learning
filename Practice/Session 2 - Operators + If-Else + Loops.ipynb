# Question 1 - Sum of Digits of a Three-Digit Number

## Problem

Write a program that accepts a **three-digit integer** and prints the sum of its digits.

> **Hint:** Use Arithmetic Operators (`%` and `//`).

### Example Input

```text
345
```

### Example Output

```text
Enter a three-digit number: 345

Sum of digits = 12
```

````python
num = int(input("Enter the three digit number: "))
a = num % 10
num = num // 10
b = num % 10
num = num // 10
c = num
print("Sum of digits =",a+b+c)
````

# Question 2 - Login System

## Problem

Create a simple login system.

Use the following credentials:

- Email : `student@python.com`
- Password : `1234`

The user will enter an email and password.

Your program should behave as follows:

- If both are correct → Print **Welcome**
- If email is correct but password is wrong → Ask the user to enter the password again.
    - If correct this time → Print **Welcome**
    - Otherwise → Print **Access Denied**
- Otherwise → Print **Invalid Email or Password**

### Example Input

```text
Enter Email
student@python.com

Enter Password
1111

Enter Password Again
1234
```

### Example Output

```text
Incorrect Password

Welcome
```

````python
email = input("Enter the Email: ")
password = int(input("Enter the Password: "))

if email == "student@python.com" and password == 1234:
    print("Welcome")
elif email == "student@python.com" and password != 1234:
    print("Password is Incorrect")
    password = int(input("Enter the Password again: "))
    if password == 1234:
        print("Welcome")
    else:
        print("Access Denied")
else:
    print("Invalid Email or Password")
````

# Question 3 - Smallest of Three Numbers

## Problem

Accept three integers from the user.

Print the smallest number **without using**:

- `min()`
- Sorting
- Lists

Use only `if`, `elif`, and `else`.

### Example Input

```text
18
7
25
```

### Example Output

```text
Smallest Number = 7
```

````python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

if a < b and a < c:
    print("Smallest Number =",a)
elif b < c:
    print("Smallest Number =",b)
else:
    print("Smallest Number =",c)
````

# Question 4 - Menu Driven Calculator

## Problem

Create a menu-driven calculator.

First, take two numbers as input.

Then display the following menu.

```text
1. Addition
2. Subtraction
3. Multiplication
4. Division
```

The user enters a choice.

Perform the corresponding operation.

If the choice is invalid, print an appropriate message.

### Example Input

```text
First Number : 15

Second Number : 5

Choice : 3
```

### Example Output

```text
Multiplication = 75
```

````python
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
print("-"*10,"Menu","-"*10)
print("   1. Addition   ")
print("   2. Subtraction   ")
print("   3. Multiplication   ")
print("   4. Division   \n")
choice = int(input("Enter the Choice: "))
if choice == 1:
    print("Addition =",a+b)
elif choice == 2:
    print("Subtraction =",a-b)
elif choice == 3:
    print("Multiplication =",a*b)
elif choice == 4:
    print("Division =",a/b)
else:
    print("Invalid Choice")
````
