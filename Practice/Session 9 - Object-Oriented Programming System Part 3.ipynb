# Question 1 — Online Shopping System (Aggregation)

Create an Online Shopping system using two classes.

### Class 1 : Address

Attributes
- House Number
- Street
- City
- State
- Pincode

Methods
- display_address()
- update_address()

---

### Class 2 : Customer

Attributes
- Customer Name
- Email
- Phone Number
- Address Object

Methods
- display_profile()
- update_profile()
- print_address()

---

### Requirements

- Customer **HAS-A** Address.
- Address must be a separate class.
- The Customer class should never directly store city, state or pincode.
- Update the address using the Address class only.

---

### Example Output

```text
Customer Name : Rahul

Email : rahul@gmail.com

Phone : 9876543210

Address

House No : 221B

Street : MG Road

City : Bengaluru

State : Karnataka

Pincode : 560001
```

````python
class Address:
    def __init__(self, house_number, street, city, state, pincode):
        self.house_no = house_number
        self.street = street
        self.city = city
        self.state = state
        self.pincode = pincode

    def display_address(self):
        print("Address")
        print(f"House No : {self.house_no}")
        print(f"Street : {self.street}")
        print(f"City : {self.city}")
        print(f"State : {self.state}")
        print(f"Pincode : {self.pincode}")

    def update_address(self, new_house_number, new_street, new_city, new_state, new_pincode):
        self.house_no = new_house_number
        self.street = new_street
        self.city = new_city
        self.state = new_state
        self.pincode = new_pincode

class Customer:

    def __init__(self, name, email, phone, address_obj):
        self.customer_name = name
        self.email = email
        self.phone_no = phone
        self.address = address_obj 

    def display_profile(self):
        print(f"Customer Name : {self.customer_name}")
        print(f"Email : {self.email}")
        print(f"Phone : {self.phone_no}")
        self.address.display_address()

    def update_profile(self, new_name, new_email, new_phone):
        self.customer_name = new_name
        self.email = new_email
        self.phone_no = new_phone

    def print_address(self):
        self.address.display_address()

my_address = Address("221B", "MG Road", "Bengaluru", "Karnataka", "560001")
customer1 = Customer("Rahul", "rahul@gmail.com", "9876543210", my_address)
customer1.display_profile()
my_address.update_address("10", "Linking Road", "Mumbai", "Maharashtra", "400050")
print("\n--- After Address Update ---")
customer1.print_address()
````

# Question 2 — University Management System (Inheritance)

Create a Parent class named `Person`.

### Parent Class

Attributes

- Name
- Age

Methods

- display_details()
- introduce()

---

Create two child classes.

### Student

Additional Attributes

- Roll Number
- Course
- CGPA

Methods

- study()

---

### Teacher

Additional Attributes

- Employee ID
- Department
- Salary

Methods

- teach()

---

### Requirements

- Student and Teacher must inherit Person.
- Do not duplicate code.
- Use inherited methods wherever possible.

---

### Example Output

```text
Name : Simran

Age : 20

Roll Number : 112

Course : B.Tech AI

CGPA : 8.9

Studying...
```

````python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_details(self):
        print(f"Name : {self.name}")
        print(f"Age : {self.age}")

    def introduce(self):
        print("Hello!")
        print(f"My name is {self.name}")
        print(f"I am {self.age} years old.")

class Student(Person):

    def __init__(self, name, age, roll_no, course, cgpa):
        super().__init__(name, age)
        self.roll_no = roll_no
        self.course = course
        self.cgpa = cgpa
    
    def study(self):
        print(f"{self.name} is studying...")

    def display_details(self):
        super().display_details()
        print(f"Roll Number : {self.roll_no}")
        print(f"Course : {self.course}")
        print(f"CGPA : {self.cgpa}")

class Teacher(Person):

    def __init__(self, name, age, id, department, salary):
        super().__init__(name, age)
        self.employee_id = id
        self.department = department
        self.salary = salary

    def teach(self):
        print(f"{self.name} is teaching...")

    def display_details(self):
        super().display_details()
        print(f"Employee ID : {self.employee_id}")
        print(f"Department : {self.department}")
        print(f"Salary : {self.salary}")

student1 = Student(name="Simran", age=20, roll_no=112, course="B.Tech AI", cgpa=8.9)
student1.display_details()
student1.study()
teacher1 = Teacher(name="Mr. Sharma", age=45, id="T-987", department="Computer Science", salary=85000)
teacher1.display_details()
teacher1.teach()
````

# Question 3 — Food Delivery App (Aggregation + Encapsulation)

Create three classes.

### Restaurant

Attributes

- Restaurant Name
- Rating

---

### Menu

Attributes

- Item Name
- Price

---

### Customer

Attributes

- Name
- Restaurant Object
- Menu Object

Methods

- place_order()
- display_bill()

---

### Requirements

- Customer HAS-A Restaurant.
- Restaurant HAS-A Menu.
- Make all attributes private.
- Access data only through getter methods.

---

### Example Output

```text
Restaurant : Pizza Hub

Item : Farmhouse Pizza

Price : 349

Order Placed Successfully.

Total Bill : ₹349
```


````python
class Menu:
    def __init__(self, item_name, price):
        self.__item_name = item_name
        self.__price = price

    def get_item_name(self):
        return self.__item_name
    
    def get_price(self):
        return self.__price
    
class Restaurant:

    def __init__(self, restaurant_name, restaurant_rate, menu_obj):
        self.__restaurant_name = restaurant_name
        self.__restaurant_rate = restaurant_rate
        self.__menu = menu_obj

    def get_restaurant_name(self):
        return self.__restaurant_name
    
    def get_restaurant_rate(self):
        return self.__restaurant_rate
    
    def get_menu(self):
        return self.__menu
    
class Customer:
    def __init__(self, name, restaurant_obj, menu_obj):
        self.__name = name
        self.__restaurant = restaurant_obj
        self.__menu = menu_obj

    def get_name(self):
        return self.__name
    
    def get_restaurant(self):
        return self.__restaurant
    
    def get_menu(self):
        return self.__menu
    
    def place_order(self):
        print(f"Restaurant : {self.get_restaurant().get_restaurant_name()}\n")
        print(f"Item : {self.get_menu().get_item_name()}\n")
        print(f"Price : {self.get_menu().get_price()}\n")
        print("Order Placed Successfully.\n")

    def display_bill(self):
        print(f"Total Bill : {self.get_menu().get_price()}")

pizza_menu = Menu(item_name="Farmhouse Pizza", price=349)
pizza_hub = Restaurant(restaurant_name="Pizza Hub", restaurant_rate=4.5, menu_obj=pizza_menu)
customer1 = Customer(name="Aarav", restaurant_obj=pizza_hub, menu_obj=pizza_menu)
customer1.place_order()
customer1.display_bill()
````

# Question 4 — Banking Application (Abstract Class)

Create an abstract class named `BankAccount`.

### Concrete Methods

- deposit(amount)
- withdraw(amount)
- check_balance()

---

### Abstract Method

```text
calculate_interest()
```

---

Create two child classes.

### SavingsAccount

Interest = 4%

---

### CurrentAccount

Interest = 0%

---

### Requirements

- Use Python's `ABC` module.
- Implement the abstract method in every child class.
- Do not create an object of the abstract class.

---

### Example Output

```text
Deposit : 10000

Current Balance : 10000

Interest : 400
```

````python
from abc import ABC, abstractmethod

class BankAccount(ABC):

    def __init__(self):
        self.balance = 0

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposit : {amount}\n")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal : {amount}\n")
        else:
            print("Insufficient funds!\n")

    def check_balance(self):
        print(f"Current Balance : {self.balance}\n")

    @abstractmethod
    def calculate_interest(self):
        pass

class SavingsAccount(BankAccount):
    def calculate_interest(self):
        # Interest = 4%
        interest = self.balance * 0.04
        print(f"Interest : {int(interest)}\n")

class CurrentAccount(BankAccount):
    def calculate_interest(self):
        # Interest = 0%
        interest = self.balance * 0
        print(f"Interest : {int(interest)}\n")

my_savings = SavingsAccount()
my_savings.deposit(10000)
my_savings.check_balance()
my_savings.calculate_interest()
````

# Question 5 — Hospital Management System (Complete OOP Challenge)

Create the following classes.

### Person (Parent)

Attributes

- Name
- Age
- Gender

---

### Doctor (Child)

Attributes

- Doctor ID
- Specialization

Methods

- diagnose()

---

### Patient (Child)

Attributes

- Patient ID
- Disease

Methods

- admit()

---

### Hospital

Attributes

- Hospital Name
- Doctor Object
- Patient Object

Methods

- admit_patient()
- discharge_patient()
- display_details()

---

### Requirements

- Use Inheritance for Doctor and Patient.
- Use Aggregation for Hospital.
- Keep all data private.
- Use getter methods where required.
- Display complete hospital details.
- Create multiple doctors and patients.

---

### Example Output

```text
Hospital : City Care Hospital

Doctor : Dr. Aman

Specialization : Cardiology

Patient : Rohit

Disease : Dengue

Patient Admitted Successfully.
```

````python
class Person:

    def __init__(self, name, age, gender):
        self.__name = name
        self.__age = age
        self.__gender = gender

    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    def get_gender(self):
        return self.__gender
    
class Doctor(Person):

    def __init__(self, name, age, gender, doctor_id, specialization):
        super().__init__(name, age, gender)
        self.__doctor_id = doctor_id
        self.__specialization = specialization

    def get_doctor_id(self):
        return self.__doctor_id
    
    def get_specialization(self):
        return self.__specialization
    
    def diagnose(self):
        print(f"Dr. {self.get_name()} is diagnosing the patient.")

class Patient(Person):

    def __init__(self, name, age, gender, patient_id, disease):
        super().__init__(name, age, gender)
        self.__patient_id = patient_id
        self.__disease = disease

    def get_patient_id(self):
        return self.__patient_id
    
    def get_disease(self):
        return self.__disease
    
    def admit(self):
        print(f"Patient {self.get_name()} is going through the admission process.")

class Hospital:

    def __init__(self, hospital_name, doctor_objects, patient_objects):
        self.__hospital_name = hospital_name
        self.__doctors = doctor_objects
        self.__patients = patient_objects

    def get_hospital_name(self):
        return self.__hospital_name
    
    def get_doctors(self):
        return self.__doctors
    
    def get_patients(self):
        return self.__patients
    
    def admit_patient(self, doctor_obj, patient_obj):
        if patient_obj not in self.__patients:
            self.__patients.append(patient_obj)
        
        print(f"Hospital : {self.get_hospital_name()}\n")
        print(f"Doctor : {doctor_obj.get_name()}\n")
        print(f"Specialization : {doctor_obj.get_specialization()}\n")
        print(f"Patient : {patient_obj.get_name()}\n")
        print(f"Disease : {patient_obj.get_disease()}\n")
        print("Patient Admitted Successfully.\n")

    def discharge_patient(self, patient_obj):
        if patient_obj in self.__patients:
            self.__patients.remove(patient_obj)
            print(f"Patient {patient_obj.get_name()} has been discharged.\n")
        else:
            print(f"Patient {patient_obj.get_name()} is not found in the records.\n")

    def display_details(self):
        print(f"--- {self.get_hospital_name()} Details ---")
        print("Doctors on Staff:")
        for doc in self.__doctors:
            print(f" - {doc.get_name()} ({doc.get_specialization()})")
        
        print("\nAdmitted Patients:")
        for pat in self.__patients:
            print(f" - {pat.get_name()} (Disease: {pat.get_disease()})")
        print("-----------------------------------")

doc1 = Doctor(name="Dr. Aman", age=45, gender="M", doctor_id="D101", specialization="Cardiology")
doc2 = Doctor(name="Dr. Sarah", age=38, gender="F", doctor_id="D102", specialization="Neurology")
pat1 = Patient(name="Rohit", age=25, gender="M", patient_id="P501", disease="Dengue")
pat2 = Patient(name="Priya", age=30, gender="F", patient_id="P502", disease="Migraine")
city_care = Hospital(hospital_name="City Care Hospital", doctor_objects=[doc1, doc2], patient_objects=[])
city_care.admit_patient(doctor_obj=doc1, patient_obj=pat1)
city_care.display_details()
````
