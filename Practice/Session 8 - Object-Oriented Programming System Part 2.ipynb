# Question 1 — 2D Point Class

Create a class named `Point`.

### Attributes
- x
- y

### Methods
- display()
- distance_from_origin()
- distance(other_point)

### Example

**Input**

```text
Point 1 : (3,4)
Point 2 : (6,8)
```

**Output**

```text
Point 1 : (3,4)
Point 2 : (6,8)

Distance from Origin (Point 1) : 5.0
Distance between Points : 5.0
```

````python
import math

class Point:

    def __init__(self,x,y):
        self.x = x
        self.y = y

    def display(self):
        return f"({self.x},{self.y})"
    
    def distance_from_original(self):
        return math.sqrt((self.x**2 + self.y**2))
    
    def distance(self,other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)
    
p1 = Point(3,4)
p2 = Point(6,8)

print(f"Point 1 : {p1.display()}")
print(f"Point 2 : {p2.display()}")
print()

print(f"Distance from Origin (Point 1) : {p1.distance_from_original()}")
print(f"Distance between Points : {p1.distance(p2)}")
````

# Question 2 — Line Class

Create a class named `Line`.

Represent a line using:

```text
Ax + By + C = 0
```

### Attributes

- A
- B
- C

### Methods

- display()
- is_point_on_line(point)

### Example

**Input**

```text
Line : x + y - 2 = 0

Point : (1,1)
```

**Output**

```text
Point lies on the line.
```

````python
class Line:
    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def display(self):
        return f"{self.A}x + {self.y}y + ({self.C})"
    
    def is_point_on_line(self,Point):
        return (self.A * Point.x + self.B * Point.y + self.C == 0)
    
line = Line(1, 1, -2)
point = Point(1,1)

if line.is_point_on_line(point):
    print("Point lies on the line.")
else:
    print("Point does not lie on the Line.")
````

# Question 3 — Shortest Distance

Using the `Point` and `Line` classes, create a method that finds the shortest distance between a point and a line.

### Example

**Input**

```text
Line : x + y - 2 = 0

Point : (1,4)
```

**Output**

```text
Shortest Distance : 2.12
```

(Round to 2 decimal places.)

````python
class Line:
    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C
        
    def shortest_distance(self, Point):
        num = abs(self.A * Point.x + self.B * Point.y + self.C)
        den = math.sqrt(self.A**2 + self.B**2)
        return round(num/den, 2)
    
line = Line(1,1,-2)
point = Point(1,4)
distance = line.shortest_distance(point)
print(f"Shortest Distance : {distance}")
````

# Question 4 — Circle Class

Create a class named `Circle`.

### Attributes

- center (Point object)
- radius

### Methods

- area()
- circumference()
- contains(point)

A point is inside the circle if its distance from the center is less than or equal to the radius.

### Example

**Input**

```text
Center : (0,0)

Radius : 5

Point : (3,4)
```

**Output**

```text
Area : 78.54

Circumference : 31.42

Point lies inside the circle.
```

````python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def distance(self,other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)
class Circle:
    def __init__(self,center,radius):
        self.center = center
        self.radius = radius

    def area(self):
        return round(math.pi*self.radius*self.radius,2)
    
    def circumference(self):
        return round(2*math.pi*self.radius,2)
    
    def point_in_circle(self,Point):
        return self.center.distance(Point) <= self.radius

center_point = Point(0, 0)
radius_value = 5
test_point = Point(3, 4)

circle = Circle(center_point, radius_value)
print(f"Area : {circle.area()}")
print(f"Circumference : {circle.circumference()}")

if circle.point_in_circle(test_point):
    print("Point lies inside the circle.")
else:
    print("Point lies outside the circle.")
````

# Question 5 — Rectangle Class Using Points

Create a class named `Rectangle`.

The rectangle is represented using two diagonal corner points.

### Attributes

- point1 (Point object)
- point2 (Point object)

### Methods

- length()
- width()
- area()
- perimeter()
- display()

### Example

**Input**

```text
Point 1 : (2,3)

Point 2 : (8,7)
```

**Output**

```text
Length : 6

Width : 4

Area : 24

Perimeter : 20
```


````python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y

    def display(self):
        return f"({self.x},{self.y})"
    
class Rectangle:
    def __init__(self, point1, point2):
        self.point1 = point1
        self.point2 = point2

    def length(self):
        return abs(self.point2.x - self.point1.x)
    
    def width(self):
        return abs(self.point2.y - self.point1.y)
    
    def area(self):
        return self.length() - self.width()
    
    def perimeter(self):
        return 2 * (self.length() + self.width())
    
    def display(self):
        print(f"Rectangle diagonally bound by Point 1 : {self.point1.display()} and Point 2 : {self.point2.display()}")

p1 = Point(2, 3)
p2 = Point(8, 7)
rect = Rectangle(p1,p2)
print(f"Length : {rect.length()}")
print() 
print(f"Width : {rect.width()}")
print()
print(f"Area : {rect.area()}")
print()
print(f"Perimeter : {rect.perimeter()}")
````
