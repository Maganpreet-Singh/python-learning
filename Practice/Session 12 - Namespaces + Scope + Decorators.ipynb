# Question 1 — LEGB Rule Demonstration

## Problem Statement

Create a program that demonstrates all four scopes of Python.

Requirements:

1. Create a global variable named `message`.
2. Create an outer function containing another variable with the same name.
3. Inside the outer function, create an inner function containing another variable with the same name.
4. Print the variable from:
   - Local Scope
   - Enclosing Scope
   - Global Scope
5. Print one built-in function (such as `len`, `max`, or `sum`) to demonstrate the Built-in Scope.
6. Explain the output using the LEGB Rule.

### Example Output

```text
Local Scope : Inner Function

Enclosing Scope : Outer Function

Global Scope : Main Program

Built-in Function : <built-in function len>
```

````python
# Global Scope
message = "Main Program"

def outer():
    # Enclosing Scope
    message = "Outer Function"

    def inner():
        # Local Scope
        message = "Inner Function"
        print("Local Scope :",message)
    inner()

    print("Enclosing Scope :",message)

outer()
print("Global Scope :",message)

# Built-in Scope
print("Built-in Function:", len)
````

# Question 2 — Global vs Local Variable Manager

## Problem Statement

Create a program that maintains a global variable named `visitor_count`.

Create the following functions:

### increase_visitors()

- Increase the global visitor count by 1.

### reset_visitors()

- Reset the visitor count to 0.

### display_visitors()

- Display the current visitor count.

Requirements:

- First, try updating the global variable **without** using the `global` keyword and observe the error.
- Then modify the program using the `global` keyword so that it works correctly.
- Display the visitor count after every operation.

### Example Output

```text
Initial Visitors : 0

Visitor Added

Current Visitors : 1

Visitor Added

Current Visitors : 2

Visitor Counter Reset

Current Visitors : 0
```

````python
visitor_count = 0

def increase_visitors():
    global visitor_count
    visitor_count += 1
    print("Visitor Added")

def reset_visitors():
    global visitor_count
    visitor_count = 0
    print("Visitor Counter Reset")

def display_visitors():
    print("Current Visitors :",visitor_count)

print("Initial Visitors :", visitor_count)
print()

increase_visitors()
display_visitors()
print()

increase_visitors()
display_visitors()
print()

reset_visitors()
display_visitors()
````

# Question 3 — Enclosing Scope using `nonlocal`

## Problem Statement

Create a program that tracks the number of completed tasks.

Create:

- An outer function named `task_manager()`
- An inner function named `complete_task()`

Requirements:

- Store the total number of completed tasks inside the outer function.
- Every time `complete_task()` is called, increase the task count by one.
- Use the `nonlocal` keyword.
- Display the updated count after every task completion.

### Example Output

```text
Task Completed

Completed Tasks : 1

Task Completed

Completed Tasks : 2

Task Completed

Completed Tasks : 3
```

````python
def task_manager():
    completed_tasks = 0

    def complete_task():
        nonlocal completed_tasks
        completed_tasks += 1
        print("Task Completed")
        print("Completed Tasks :", completed_tasks)
        print()

    complete_task()
    complete_task()
    complete_task()


task_manager()
````

# Question 4 — First-Class Functions

## Problem Statement

Create three mathematical functions.

```text
Addition

Subtraction

Multiplication
```

Create another function named

```text
calculator()
```

Requirements:

- Pass one function as an argument to another function.
- Store functions inside a list.
- Store one function inside another variable.
- Call the function using the new variable.
- Demonstrate that functions are first-class citizens in Python.

### Example Output

```text
Addition Result : 25

Subtraction Result : 15

Multiplication Result : 100

Calling Function Stored Inside Variable

Addition Result : 25
```


````python
def addition(a, b):
    print("Addition Result :",a + b)

def subtraction(a, b):
    print("Subtraction Result :",a - b)

def multiplication(a, b):
    print("Multiplication Result :",a * b)

def calculator(operation, a, b):
    operation(a, b)

calculator(addition, 20, 5)
print()

calculator(subtraction, 20, 5)
print()

calculator(multiplication, 20, 5)
print()

operations = [addition, subtraction, multiplication]
print("Calling Functions from List")
for operation in operations:
    operation(20, 5)

print()
new_function = addition
print("Calling Function Stored Inside Variable")
new_function(20, 5)
````

# Question 5 — User-defined Decorator

## Problem Statement

Create a decorator named

```python
execution_logger
```

The decorator should:

- Print `"Function Started"` before executing the original function.
- Print `"Function Finished"` after execution.
- Return the original result.

Create the following functions and decorate them:

- greet(name)
- square(number)
- student_details(name, age)

Requirements:

- Use `@execution_logger`
- Decorate all three functions.
- Ensure the original return value is preserved.
- Demonstrate the decorator with different inputs.

### Example Output

```text
Function Started

Hello Rahul

Function Finished

----------------------

Function Started

Square = 64

Function Finished

----------------------

Function Started

Name : Aman

Age : 20

Function Finished
```


````python
def execution_logger(func):
    def wrapper(*args, **kwargs):
        print("Function Started")

        result = func(*args, **kwargs)

        print("Function Finished")
        return result
    
    return wrapper

@execution_logger
def greet(name):
    print(f"Hello {name}")

@execution_logger
def square(number):
    result = number ** 2
    print("Square =", result)
    return result

@execution_logger
def student_details(name, age):
    print("Name :", name)
    print("Age :", age)

greet("Rahul")

print("-" * 22)

square_result = square(8)
print("Returned Value :", square_result)

print("-" * 22)

student_details("Aman", 20)
````
