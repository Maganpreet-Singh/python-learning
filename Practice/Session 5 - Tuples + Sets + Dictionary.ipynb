# Question 2 — Set Practice

Create the following set.

```python
A = {10,20,30,40}
```

Perform the following tasks.

1. Add 50.
2. Add 60, 70 and 80 together.
3. Remove 20.
4. Discard 100.
5. Remove one random element using pop().
6. Print the length of the set.
7. Check whether 30 exists.
8. Check whether 200 is not present.

### Example Output

```text
Initial Set
{10, 20, 30, 40}

After add()
{10, 20, 30, 40, 50}

After update()
{10, 20, 30, 40, 50, 60, 70, 80}

After remove()
{10, 30, 40, 50, 60, 70, 80}

After discard()
{10, 30, 40, 50, 60, 70, 80}

Length
7

True

True
```


````python
A = {10,20,30,40}
print(A)
A.add(50)
print(A)
A.update([60, 70, 80])
print(A)
A.remove(20)
print(A)
A.discard(100)
print(A)
print(len(A))
print(30 in A)
print(200 not in A)
````

# Question 4 — Dictionary Practice

Create the following dictionary.

```python
student = {
    "name": "Rahul",
    "age": 20,
    "course": "CSE",
    "marks": 92
}
```

Perform the following tasks.

1. Print the student's name.
2. Print the student's marks.
3. Add a new key `city` with value `"Delhi"`.
4. Update marks to `95`.
5. Remove the `age` key.
6. Print all keys.
7. Print all values.
8. Print all key-value pairs.

### Example Output

```text
Rahul

92

{
'name': 'Rahul',
'course': 'CSE',
'marks': 95,
'city': 'Delhi'
}

dict_keys(['name','course','marks','city'])

dict_values(['Rahul','CSE',95,'Delhi'])

dict_items([
('name','Rahul'),
('course','CSE'),
('marks',95),
('city','Delhi')
])
```


````python
student = {
    "name": "Rahul",
    "age": 20,
    "course": "CSE",
    "marks": 92
}

print(student["name"])
print(student["marks"])
student["city"] = "Delhi"
print(student)
student["marks"] = 95
print(student)
student.pop("age")
print(student)
print(student.keys())
print(student.values())
print(student.items())
````

# Question 5 — Comprehensive Challenge

Create the following data.

```python
student = {
    "name": "Aman",
    "age": 19,
    "subjects": ("Python", "AI", "Math"),
    "skills": {"Python", "SQL"},
    "marks": {
        "Python": 90,
        "AI": 95,
        "Math": 88
    }
}
```

Perform the following tasks.

1. Print the student's name.
2. Print the second subject.
3. Print the AI marks.
4. Add `"Git"` to the skills set.
5. Add a new subject `"Cloud"` to the marks dictionary with value `91`.
6. Update Python marks to `96`.
7. Print all subjects.
8. Print all skills.
9. Print all marks.
10. Print the complete dictionary.

### Example Output

```text
Aman

AI

95

{'Python', 'SQL', 'Git'}

{
'Python':96,
'AI':95,
'Math':88,
'Cloud':91
}

('Python','AI','Math')

```

````python
student = {
    "name": "Aman",
    "age": 19,
    "subjects": ("Python", "AI", "Math"),
    "skills": {"Python", "SQL"},
    "marks": {
        "Python": 90,
        "AI": 95,
        "Math": 88
    }
}

print(student["name"])
print(student["subjects"][1])
print(student["marks"]["AI"])
student["skills"].add("Git")
print(student["skills"])
student["marks"]["Cloud"] = 91
print(student["marks"])
student["marks"]["Python"] = 96
print(student["marks"])
print(student["subjects"])
print(student["skills"])
print(student["marks"])
print(student)
````
