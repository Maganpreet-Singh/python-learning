# Question 1 — Smart Calculator

## Problem Statement

Create a calculator that performs one of the following operations:

- Addition
- Subtraction
- Multiplication
- Division

The user enters:

1. First number
2. Operator (+, -, *, /)
3. Second number

The program should:

- Perform the calculation if inputs are valid.
- Handle division by zero.
- Handle invalid numbers.
- Handle invalid operators.
- Use `try`, `except`, `else`, and `finally`.

### Example Input 1

```
Enter first number:
20

Enter operator:
/

Enter second number:
4
```

### Example Output 1

```
Result = 5.0

Program executed successfully.

Calculator Closed.
```

---

### Example Input 2

```
Enter first number:
15

Enter operator:
/

Enter second number:
0
```

### Example Output 2

```
Error:
Cannot divide by zero.

Calculator Closed.
```

---

### Example Input 3

```
Enter first number:
abc
```

### Example Output 3

```
Invalid number entered.

Calculator Closed.
```


````python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    operator = input("Enter operator (+, -, *, /): ")

    if operator == "+":
        result = num1 + num2
    elif operator == "-":
        result = num1 - num2
    elif operator == "*":
        result = num1 * num2
    elif operator == "/":
        if num2 == 0:
            raise ZeroDivisionError("Cannot divided by zero.")
        result = num1 / num2
    else:
        raise ValueError("Invalid operator.")

except ValueError as e:
    if "could not convert string to float" in str(e):
        print("Invalid number entered.")
    else:
        print("Error:")
        print(e)

except ZeroDivisionError as e:
    print("Error:")
    print(e)

else:
    print(f"Result = {result}")
    print("\nProgram executed successfully.")

finally:
    print("\nCalculator Closed.")
````

# Question 2 — Bank Withdrawal System

## Problem Statement

A bank account contains ₹10,000.

Ask the user to enter the withdrawal amount.

Rules:

- Amount cannot be negative.
- Amount cannot be zero.
- Amount cannot exceed account balance.
- If withdrawal is successful, update the balance.
- Use `raise` to generate appropriate exceptions.
- Handle exceptions using `try-except`.

### Example Input 1

```
Balance = 10000

Withdraw Amount:
3500
```

### Example Output 1

```
Transaction Successful

Remaining Balance = 6500
```

---

### Example Input 2

```
Withdraw Amount:
15000
```

### Example Output 2

```
Error:
Insufficient Balance
```

---

### Example Input 3

```
Withdraw Amount:
-500
```

### Example Output 3

```
Error:
Withdrawal amount cannot be negative.
```


````python
balance = 10000

try:
    print(f"Balance = {balance}")
    amount = float(input("Withdraw Amount: "))

    if amount < 0:
        raise ValueError("Withdrawal amount cannot be negative.")
    elif amount == 0:
        raise ValueError("Withdrawal amount cannot be zero.")
    elif amount > balance:
        raise ValueError("Insufficient Balance")
    
    balance -= amount

except ValueError as e:
    print("Error:")
    print(e)

else:
    print("Transaction Successful")
    print(f"\nRemaining Balance = {balance}")

finally:
    print("\nThank you for banking with us.")
````

# Question 3 — Student Result Validator

## Problem Statement

Take marks of 5 students.

Rules:

- Marks must be between 0 and 100.
- If marks are outside the range, raise an exception.
- Store only valid marks.
- After all inputs:
  - Display all marks.
  - Display highest mark.
  - Display lowest mark.
  - Display average marks.
- Use `try`, `except`, `raise`, `else`, and `finally`.

### Example Input

```
Student 1 : 89
Student 2 : 75
Student 3 : 101
Student 4 : 65
Student 5 : -8
```

### Example Output

```
Student 3:
Marks must be between 0 and 100.

Student 5:
Marks must be between 0 and 100.

Valid Marks

89
75
65

Highest = 89

Lowest = 65

Average = 76.33
```


````python
marks = []

for i in range(1,6):
    try:
        mark = float(input(f"Student {i}: "))

        if mark < 0 or mark > 100:
            raise ValueError("Marks must be between 0 and 100.")
        
    except ValueError as e:
        print(f"\nStudent {i}:")
        print(e)

    else:
        marks.append(mark)

    finally:
        pass

if marks:
    print("\nValid Marks")
    for mark in marks:
        print(mark)

    print(f"\nHighest = {max(marks)}")
    print(f"\nLowest = {min(marks)}")
    print(f"\nAverage = {round(sum(marks) / len(marks),2)}")
else:
    print("\nNo Valid marks were entered.")
````

# Question 4 — File Reader with Exception Handling

## Problem Statement

Ask the user for a file name.

The program should:

- Open the file.
- Read its contents.
- Display the total number of lines.
- Display the total number of words.
- Display the total number of characters.

Handle:

- FileNotFoundError
- PermissionError
- Any other unexpected exception

Use:

- try
- multiple except blocks
- else
- finally

### Example Input 1

```
Enter file name:

notes.txt
```

### Example Output 1

```
File Opened Successfully

Total Lines : 18

Total Words : 245

Total Characters : 1430

File Closed Successfully
```

---

### Example Input 2

```
Enter file name:

python.txt
```

### Example Output 2

```
Error:
File not found.

File Closed Successfully
```


````python
file = None

try:
    file_name = input("Enter file name: ")

    file = open(file_name, "r")

    content = file.read()

    lines = content.splitlines()
    words = content.split()
    characters = len(content)

except FileNotFoundError:
    print("Error:")
    print("File not found.")

except PermissionError:
    print("Error:")
    print("Permission denied.")

except Exception as e:
    print("Error:")
    print(e)

else:
    print("File Opened Successfully")
    print(f"\nTotal Lines : {len(lines)}")
    print(f"\nTotal Words : {len(words)}")
    print(f"\nTotal Characters : {characters}")

finally:
    if file:
        file.close()
    print("\nFile Closed Successfully")
````

# Question 5 — College Admission System (Custom Exception)

## Problem Statement

Create your own custom exception named

```
InvalidAgeError
```

Ask the user for:

- Name
- Age
- Percentage

Rules:

- Age must be between 17 and 25.
- Percentage must be between 0 and 100.
- If any rule is violated, raise the appropriate exception.
- If all inputs are valid, print the admission details.
- Use a custom exception for age validation and built-in exceptions where appropriate.

### Example Input 1

```
Name:
Rahul

Age:
19

Percentage:
87
```

### Example Output 1

```
Admission Successful

Name : Rahul

Age : 19

Percentage : 87%
```

---

### Example Input 2

```
Name:
Aman

Age:
15

Percentage:
92
```

### Example Output 2

```
InvalidAgeError:
Student age must be between 17 and 25.
```

---

### Example Input 3

```
Name:
Priya

Age:
20

Percentage:
120
```

### Example Output 3

```
Error:
Percentage must be between 0 and 100.
```

````python
class InvalidAgeError(Exception):
    pass

try:
    name = input("Name: ")
    age = int(input("Age: "))
    if age < 17 or age > 25:
        raise InvalidAgeError("Student age must be between 17 and 25.")
    
    percentage = float(input("Percentage: "))
    if percentage < 0 or percentage > 100:
        raise ValueError("Percentage must be between 0 and 100.")
    
except InvalidAgeError as e:
    print("InvalidAgeError:")
    print(e)

except ValueError as e:
    print("Error:")
    print(e)

else:
    print("\nAdmission Successful")
    print(f"\nName : {name}")
    print(f"\nAge : {age}")
    print(f"\nPercentage : {percentage}%")

finally:
    print("\nAdmission Process Completed.")
````
