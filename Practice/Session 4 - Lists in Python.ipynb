# Question 3 — Practice Positive, Negative Indexing and Slicing

Create the following list.

```python
numbers = [5,10,15,20,25,30,35,40,45,50]
```

Print the following.

### Positive Indexing

- First element
- Fourth element
- Last element

### Negative Indexing

- Last element
- Third last element
- Fifth last element

### Slicing

1. First five elements
2. Last four elements
3. Elements from index 2 to index 7
4. Every second element
5. Every third element
6. Reverse the list

### Example Output

```text
5
20
50
50
40
30
[5, 10, 15, 20, 25]
[35, 40, 45, 50]
[15, 20, 25, 30, 35, 40]
```

# Question 4 — Add Items into Lists

Create the following list.

```python
languages = ["Python", "Java"]
```

Perform the following operations one by one.

1. Add `"C++"` using `append()`
2. Add `"JavaScript"` and `"Go"` together using `extend()`
3. Insert `"C"` at index 1
4. Insert `"Rust"` at the beginning
5. Insert `"Swift"` before the last element

Print the list after every operation.

### Example Output

```text
['Python', 'Java', 'C++']

['Python', 'Java', 'C++', 'JavaScript', 'Go']

['Python', 'C', 'Java', 'C++', 'JavaScript', 'Go']

['Rust', 'Python', 'C', 'Java', 'C++', 'JavaScript', 'Go']

['Rust', 'Python', 'C', 'Java', 'C++', 'JavaScript', 'Swift', 'Go']
```

````python
languages = ["Python", "Java"]

languages.append("C++")
print(languages)
languages.extend(["JavaScript", "Go"])
print(languages)
languages.insert(1,"C")
print(languages)
languages.insert(0,"Rust")
print(languages)
languages.insert(-1,"Swift")
print(languages)
````

# Question 5 — Comprehensive List Challenge

Create the following nested list.

```python
college = [
    "ABC University",
    [
        ["CSE", 120],
        ["AI", 90],
        ["ECE", 60]
    ],
    [
        ["Rahul", 85],
        ["Aman", 92],
        ["Priya", 88]
    ]
]
```

## Part A

Print using indexing only.

- University Name
- AI
- 90
- Priya
- 88

## Part B

Using slicing,

Print

- First two branches
- Last two students

## Part C

Perform the following operations.

1. Add branch `"IT"` with 75 seats.
2. Add student `"Neha"` with marks 95.
3. Insert branch `"Mechanical"` with 110 seats at index 1.
4. Print the final nested list.

### Example Output

```text
ABC University

AI

90

Priya

88

[['CSE', 120], ['AI', 90]]

[['Aman', 92], ['Priya', 88]]

[['Priya', 88], ['Aman', 92], ['Rahul', 85]]

[
 'ABC University',
 [
  ['CSE',120],
  ['Mechanical',110],
  ['AI',90],
  ['ECE',60],
  ['IT',75]
 ],
 [
  ['Rahul',85],
  ['Aman',92],
  ['Priya',88],
  ['Neha',95]
 ]
]
```

````python
college = [
    "ABC University",
    [
        ["CSE", 120],
        ["AI", 90],
        ["ECE", 60]
    ],
    [
        ["Rahul", 85],
        ["Aman", 92],
        ["Priya", 88]
    ]
]
# Part A
print(college[0])
print(college[1][1][0])
print(college[1][1][1])
print(college[2][2][0])
print(college[2][2][1])
# Part B
print(college[1][:2])
print(college[2][-2:])
print(college[2][::-1])
# Part C
college[1].append(["IT", 75])
college[2].append(["Neha", 95])
college[1].insert(1,["Mechanical", 110])
print(college)
````

# Bonus Challenge (No Loops)

Create the following 3D list.

```python
cube = [
    [
        [11,12],
        [13,14]
    ],
    [
        [15,16],
        [17,18]
    ]
]
```

Using indexing only, print

- 11
- 14
- 15
- 18
- [13,14]
- [15,16]
- Entire second layer

### Example Output

```text
11
14
```

````python
cube = [
    [
        [11,12],
        [13,14]
    ],
    [
        [15,16],
        [17,18]
    ]
]

print(cube[0][0][0])
print(cube[0][1][1])
print(cube[1][0][0])
print(cube[1][1][1])
print(cube[0][1])
print(cube[1][0])
print(cube[1])
````
