# Question 1 — Student Report File Manager

## Objective

Create a program that stores student records inside a text file.

### Step 1

Create a file named

```
students.txt
```

Store the following records.

```
Rahul,101,91
Simran,102,87
Aman,103,79
Priya,104,95
Rohan,105,82
```

---

### Step 2

Read the complete file.

Display

```
----- Student Records -----

Rahul,101,91
Simran,102,87
Aman,103,79
Priya,104,95
Rohan,105,82
```

---

### Step 3

Append the following record.

```
Karan,106,89
```

---

### Step 4

Read the file again and display the updated contents.

---

### Requirements

- Use write()
- Use append mode
- Use read()
- Use Context Manager (`with`)

````python
# Step 1:
records = """Rahul,101,91
Simran,102,87
Aman,103,79
Priya,104,95
Rohan,105,82"""

with open("students.txt", "w") as file:
    file.write(records)
````

````python
# Step 2:
print("----- Student Records -----\n")
with open("students.txt",'r') as file:
    print(file.read())
````

````python
# Step 3:
new_record = "\nKaran,106,89"
with open("students.txt","a") as file:
    file.write(new_record)
````

````python
# Step 4:
print("\n----- Updated Student Records -----\n")
with open("students.txt", "r") as file:
    updated_content = file.read()
    print(updated_content)
````

# Question 2 — Employee Attendance System

Create a file named

```
attendance.txt
```

Store attendance for 7 employees.

Example

```
Rahul - Present
Aman - Absent
Priya - Present
Simran - Present
Karan - Absent
Rohit - Present
Neha - Present
```

---

Perform the following tasks.

### Part A

Read one line at a time using

```
readline()
```

Display every line separately.

---

### Part B

Read the entire file using

```
readlines()
```

Display

- Total Employees
- Present Employees
- Absent Employees

---

### Expected Output

```
Total Employees : 7

Present : 5

Absent : 2
```

---

### Concepts Covered

- readline()
- readlines()
- String Processing
- File Reading

````python
records = """Rahul - Present
Aman - Absent
Priya - Present
Simran - Present
Karan - Absent
Rohit - Present
Neha - Present"""

with open("attendance.txt","w") as f:
    f.write(records)
````

````python
with open("attendance.txt","r") as f:
    while True:
        line = f.readline()
        if not line:
            break
        print(line.strip())
````

````python
with open("attendance.txt","r") as f:
    all_lines = f.readlines()

total_employees = len(all_lines)
present_count = 0
absent_count = 0

for line in all_lines:
    if "Present" in line:
        present_count += 1
    elif "Absent" in line:
        absent_count += 1

print(f"Total Employees : {total_employees}")
print(f"Present : {present_count}")
print(f"Absent : {absent_count}")
````

# Question 3 — Personal Diary Application

Create a file named

```
diary.txt
```

The program should perform the following operations.

### Menu

```
1. Write New Diary Entry

2. Read Entire Diary

3. Add New Entry

4. Exit
```

---

Requirements

If the user selects

### Option 1

Create a new diary.

Previous content should be removed.

---

### Option 2

Display the complete diary.

---

### Option 3

Append a new diary entry without deleting previous entries.

---

### Option 4

Terminate the program.

---

Example Output

```
Today's Entry

Today I completed Python File Handling.

Entry Saved Successfully.

-------------------

Previous Entries

Today I completed Python File Handling.

Tomorrow I will practice more.
```

---

### Concepts Covered

- write()
- append()
- read()
- File Modes
- Context Manager

````python
import os
while True:
    print("\n" + "="*30)
    print("     PERSONAL DIARY MENU     ")
    print("="*30)
    print("1. Write New Diary Entry (Overwrite)")
    print("2. Read Entire Diary")
    print("3. Add New Entry (Append)")
    print("4. Exit")
    print("="*30)

    choice = input("Enter your choice (1-4): ").strip()

    if choice == "1":
        entry = input("\nEnter your new diary entry:\n>")
        with open("diary.txt","w") as file:
            file.write(entry + "\n")
        print("\n[Success] New diary created and entry saved!")

    elif choice == "2":
        print("\n----- Personal Diary Content -----")
        try:
            with open("diary.txt","r") as file:
                content = file.read()
                if content.strip() == "":
                    print("(The diary is currently empty)")
                else:
                    print(content, end="")
        except FileNotFoundError:
            print("(No diary file found yet. Write a new entry to create one!)")
        print("-"*34)

    elif choice == "3":
        file_exists = os.path.exists("diary.txt")
        entry = input("\nEnter the entry to add:\n")
        with open("diary.txt","a") as file:
            file.write(entry + "\n")
        print("\n[Success] Entry appended successfully!")

    elif choice == "4":
        print("\nExiting Diary Application. Goodbye!")
        break

    else:
        print("\n[Error] Invalid choice! Please enter a number between 1 and 4.")
````

# Question 4 — File Pointer Practice

Create a file named

```
story.txt
```

Store a paragraph of at least 150 characters.

Perform the following tasks.

### Task 1

Read the first 20 characters.

---

### Task 2

Display the current cursor position using

```
tell()
```

---

### Task 3

Move the cursor to

- Beginning
- 15th character
- 40th character

using

```
seek()
```

Read 10 characters after every movement.

---

### Example Output

```
First 20 Characters

Python is a power

Current Position

20

Move to Beginning

Python is

Move to Position 15

powerful p

Move to Position 40

used for D
```

---

### Concepts Covered

- seek()
- tell()
- read()


````python
story_content = (
    "Python is a powerful programming language widely used for Data Science, "
    "machine learning, web development, and automation. It is known for its "
    "simplicity, readability, and great community support worldwide."
)
with open("story.txt","w") as file:
    file.write(story_content)

with open("story.txt","r") as file:
    print("First 20 Characters\n")
    first_20 = file.read(20)
    print(first_20)
    print("\n","-"*20,"\n")

    print("Current Position\n")
    position = file.tell()
    print(position)
    print("\n","-"*20,"\n")

    print("Move to Beginning\n")
    file.seek(0)
    print(file.read(10))
    print("\n-------------------\n")

    print("Move to Position 15\n")
    file.seek(15)
    print(file.read(10))
    print("\n-------------------\n")

    print("Move to Position 40\n")
    file.seek(40)
    print(file.read(10))
````

# Question 5 — Large File Reader

Create a file named

```
large_data.txt
```

Write at least

```
1000 lines
```

inside the file.

Example

```
Record 1

Record 2

Record 3

...

Record 1000
```

---

Now read the file

NOT all at once.

Read only

```
50 characters
```

at a time.

After every chunk display

```
-----------------------
Chunk Number : 1
-----------------------

<50 characters>

Current Cursor Position : 50
```

Continue until the entire file is read.

At the end display

```
File Reading Completed.
```

---

### Requirements

- Use read(size)
- Use tell()
- Use loop
- Use Context Manager

````python
with open("large_data.txt","w") as file:
    for i in range(1,1001):
        file.write(f"Record {i}\n")

with open("large_data.txt","r") as file:
    chunk_number = 1

    while True:
        chunk = file.read(50)
        
        if not chunk:
            break
        
        print("-----------------------")
        print(f"Chunk Number : {chunk_number}")
        print("-----------------------")
        print()
        print(chunk, end="")  # end="" prevents adding an extra newline since chunk already has them
        print("\n")
        print(f"Current Cursor Position : {file.tell()}")
        print()
        
        chunk_number += 1

print("File Reading Completed.")
````
