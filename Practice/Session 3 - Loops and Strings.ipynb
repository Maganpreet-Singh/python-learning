## Question 1: Print All Ordered Pairs

Write a program that takes an integer `n` and prints all ordered pairs `(i, j)` where both `i` and `j` range from `1` to `n`.

### Example Input
```
3
```

### Example Output
```
(1,1)
(1,2)
(1,3)
(2,1)
(2,2)
(2,3)
(3,1)
(3,2)
(3,3)
```

````python
n = int(input("Enter the number: "))
for i in range(1,n+1):
    for j in range(1,n+1):
        print((i,j))
````

## Question 2: Number Triangle Pattern

Write a program that prints the following pattern.

### Example Input
```
5
```

### Example Output
```
1
12
123
1234
12345
```

````python
n = int(input("Enter the number: "))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end="")
    print()
````

## Question 3: Palindrome Number Pattern

Write a program to print the following pattern.

### Example Input
```
5
```

### Example Output
```
1
121
12321
1234321
123454321
```

````python
n = int(input("Enter the number: "))
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end="")
    for k in range(i-1,0,-1):
        print(k,end="")
    print()
````

## Question 4: Find the First Number Divisible by 7

Given two integers `start` and `end`, print the first number between them (inclusive) that is divisible by `7`. Stop searching immediately after finding it.

### Example Input
```
20
50
```

### Example Output
```
21
```

````python
start = int(input("Enter the starting point: "))
end = int(input("Enter the ending point: "))

for i in range(start,end):
    if i % 7 == 0:
        print(i)
        break
````

## Question 5: Print Numbers Except Multiples of 3

Write a program that prints all numbers from `1` to `n`, but skips numbers that are divisible by `3`.

### Example Input
```
12
```

### Example Output
```
1
2
4
5
7
8
10
11
```

````python
n = int(input("Enter the number: "))
for i in range(1,n+1):
    if i % 3 == 0:
        continue
    print(i)
````

## Question 8: Count Uppercase, Lowercase, Digits and Special Characters

Write a program that counts:
- Uppercase letters
- Lowercase letters
- Digits
- Special characters

### Example Input
```
PyTh0n@123
```

### Example Output
```
Uppercase Letters : 2
Lowercase Letters : 3
Digits            : 4
Special Characters: 1
```

````python
s = input("Enter the string: ")
upper = 0
lower = 0
digit = 0
special = 0
for char in s:
    if char.isupper():
        upper += 1
    elif char.islower():
        lower += 1
    elif char.isdigit():
        digit += 1
    else:
        special += 1
print(f"Uppercase Letters : {upper}")
print(f"Lowercase Letters : {lower}")
print(f"Digits            : {digit}")
print(f"Special Characters: {special}")
````

## Question 10: Character Frequency

Write a program that takes a string and prints how many times each character appears.

### Example Input
```
banana
```

### Example Output
```
b : 1
a : 3
n : 2
```

````python
s = input("Enter the string: ")
freq = {}
for char in s:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

for char,count in freq.items():
    print(f"{char} : {count}")
````
