# Question 1 — Student Grade Function

Write a function named `student_grade()`.

The function should:

- Accept a student's name and marks.
- Return:
  - "A" if marks are 90 or above
  - "B" if marks are between 75 and 89
  - "C" if marks are between 60 and 74
  - "Fail" otherwise

Call the function for different students using both **positional** and **keyword** arguments.

### Example Input

```text
Rahul
92
```

### Example Output

```text
Rahul : Grade A
```

````python
def student_grade(name, marks):
    if isinstance(name, str) and isinstance(marks, (int, float)):
        if marks >= 90:
            return(f"{name} : Grade A")
        elif 75 <= marks < 90:
            return(f"{name} : Grade B")
        elif 60 <= marks < 75:
            return(f"{name} : Grade C")
        else:
            return(f"{name} : Fail")
    else:
        return "Invalid Input"
    
print(student_grade("Rahul",92))
print(student_grade("Aman", 88.5))     
print(student_grade("Priya", "Eighty"))  
print(student_grade(123, 70))
print(student_grade(name="Priya", marks=85))
print(student_grade(marks=45, name="Neha"))
````

# Question 2 — Electricity Bill Calculator

Create a function named `electricity_bill()`.

Parameters

- units
- rate (default = 8)

The function should calculate

```
Bill = units × rate
```

Tasks

1. Call the function by passing both parameters.
2. Call the function using only units.
3. Call the function using keyword arguments.
4. Print the returned bill.

### Example Input

```text
Units = 120
Rate = 10
```

### Example Output

```text
Total Bill = 1200
```

Second Call

```text
Units = 120
```

Output

```text
Total Bill = 960
```

````python
def electricity_bill(units, rate = 8):
    if isinstance(units, (int, float)) and isinstance(rate, (int, float)):
        Bill = units * rate
        return f"Total Bill = {Bill}"
    else:
        return "Invalid Input"

print(electricity_bill(120,10))
print(electricity_bill(120))
print(electricity_bill(rate= 75, units= 7584.5))
````

# Question 3 — Shopping Cart using *args

Write a function named `shopping_cart()`.

The function should accept any number of item prices using `*args`.

It should print

- Total number of items
- Total bill
- Highest priced item
- Lowest priced item
- Average price

### Example Input

```text
250
400
120
80
650
```

### Example Output

```text
Items = 5

Total Bill = 1500

Highest Price = 650

Lowest Price = 80

Average Price = 300.0
```

````python
def shopping_cart(*items):
    if all(isinstance(item, (int, float)) for item in items):
        print(f"Items = {len(items)}")
        bill = 0
        for i in items:
            bill += i
        print(f"Total Bill = {bill}")
        print(f"Highest Price = {max(items)}")
        print(f"Lowest Price = {min(items)}")
        print(f"Average Price = {bill/len(items)}")
    else:
        print("Invalid Input")
    
shopping_cart(250,400,120,800,650,4648,87845,7784,87)
````

# Question 4 — Student Information using **kwargs

Write a function named `student_details()`.

The function should accept any number of keyword arguments using `**kwargs`.

Display all key-value pairs neatly.

Call the function with

- name
- age
- course
- city
- CGPA
- semester

### Example Call

```python
student_details(
    name="Rahul",
    age=20,
    course="CSE",
    city="Delhi",
    CGPA=8.75,
    semester=3
)
```

### Example Output

```text
Student Details

name : Rahul
age : 20
course : CSE
city : Delhi
CGPA : 8.75
semester : 3
```

````python
def student_details(**details):
    print("Student Details\n")
    for key,value in details.items():
        print(f"{key} : {value}")

student_details(
    name="Rahul",
    age=20,
    course="CSE",
    city="Delhi",
    CGPA=8.75,
    semester=3
)
````

# Question 5 — Comprehensive Function Challenge

Create a function named `employee_salary()`.

Requirements

- Add a proper docstring.
- Parameters
    - name
    - basic_salary
    - bonus (default = 5000)
    - tax (default = 10)

Calculate

```
Gross Salary = Basic Salary + Bonus

Tax Amount = Gross Salary × tax / 100

Net Salary = Gross Salary − Tax Amount
```

The function should **return** the net salary.

Perform the following tasks.

1. Call the function using positional arguments.
2. Call the function using keyword arguments.
3. Call the function without passing bonus.
4. Print the function's documentation using `__doc__`.

### Example Input

```text
Name = Aman

Basic Salary = 50000

Bonus = 8000

Tax = 10
```

### Example Output

```text
Employee : Aman

Gross Salary : 58000

Tax : 5800

Net Salary : 52200

This function calculates the net salary of an employee.
```

````python
def employee_salary(name, basic_salary, bonus = 5000, tax = 10):
    if isinstance(name, str) and isinstance(basic_salary, (int, float)) and isinstance(bonus, (int, float)) and isinstance(tax, (int, float)):
        gross_salary = basic_salary + bonus
        tax_amoount = gross_salary * (tax / 100)
        net_salary = gross_salary - tax_amoount
        print(f"Employee : {name}")
        print(f"Gross Salary : {gross_salary}")
        print(f"Tax : {tax_amoount}")
        print(f"Net Salary : {net_salary}")
    else:
        print("Invalid Input")

employee_salary(name="Aman", basic_salary= 50000, bonus= 8000, tax= 10)
````
