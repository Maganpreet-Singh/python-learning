# Question 1 - Student Information Card

## Problem

Create variables for the following details:

- Name
- Age
- College
- Branch
- Semester
- CGPA
- Is Student (Boolean)

Print the information in a neatly formatted student profile.

After printing the profile, print the data type of every variable using the `type()` function.

### Example Output

```
------------------------------
      STUDENT PROFILE
------------------------------

Name      : Rahul Sharma
Age       : 20
College   : ABC University
Branch    : CSE
Semester  : 3
CGPA      : 8.75
Is Student: True

Data Types

Name       -> <class 'str'>
Age        -> <class 'int'>
College    -> <class 'str'>
Branch     -> <class 'str'>
Semester   -> <class 'int'>
CGPA       -> <class 'float'>
Is Student -> <class 'bool'>
```

````python
print("------------------------------")
print("        STUDENT PROFILE       ")
print("------------------------------\n")

name = input("Enter your name: ")
age = int(input("Enter your age: "))
college = input("Enter your college name: ")
branch = input("Enter your branch: ")
semester = int(input("Enter the current semester: "))
cgpa = float(input("Enter your CGPA: "))
is_student = bool(input("Are you student?: "))
print("Name      : ",name)
print("Age       : ",age)
print("College   : ",college)
print("Branch    : ",branch)
print("Semester  : ",semester)
print("CGPA      : ",cgpa)
print("Is Student: ",is_student,"\n")
print("Data Types\n")
print("Name       -> ",type(name))
print("Age        -> ",type(name))
print("College    -> ",type(name))
print("Branch     -> ",type(name))
print("Semester   -> ",type(name))
print("CGPA       -> ",type(name))
print("Is Student -> ",type(name))
````

# Question 2 - Basic Calculator Report

## Problem

Create two variables containing any two integer values.

Print the following operations:

- Addition
- Subtraction
- Multiplication
- Division
- Floor Division
- Modulus
- Exponent

### Example

**Sample Values**

```
First Number  : 25
Second Number : 6
```

### Example Output

```
----------- CALCULATOR -----------

First Number  : 25
Second Number : 6

Addition        : 31
Subtraction     : 19
Multiplication  : 150
Division        : 4.1666666667
Floor Division  : 4
Modulus         : 1
Exponent        : 244140625
```

````python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("-"*10,"CALCULATOR","-"*10,"\n")
print("First Number  : ",a)
print("Second Number : ",b,"\n")
print("Addition        :",a+b)
print("Subtraction     :",a-b)
print("Multiplication  :",a*b)
print("Division        :",a/b)
print("Floor Division  :",a//+b)
print("Modulus         :",a%b)
print("Exponent        :",a**b)
````

# Question 5 - Mini Employee Database

## Problem

Create:

- One Dictionary containing employee details
- One List containing five programming languages
- One Tuple containing three project names
- One Set containing duplicate numbers
- One Boolean variable indicating whether the employee is active

Print all variables neatly.

Also print the data type of each variable.

### Example Output

```
=========== EMPLOYEE DATABASE ===========

Employee

{'ID':101, 'Name':'Rahul', 'Department':'AI', 'Salary':60000}

Programming Languages

['Python', 'Java', 'C++', 'SQL', 'JavaScript']

Projects

('ChatBot', 'Weather App', 'Portfolio')

Unique Numbers

{10, 20, 30, 40}

Employee Active

True

---------------------------------

Data Types

Employee Dictionary : <class 'dict'>
Programming Languages : <class 'list'>
Projects : <class 'tuple'>
Unique Numbers : <class 'set'>
Employee Active : <class 'bool'>
```

````python
print("="*10,"EMPLOYEE DATABASE","="*10,"\n")
employee = {'ID':101, 'Name':'Rahul', 'Department':'AI', 'Salary':60000}
programming_language = ['Python', 'Java', 'C++', 'SQL', 'JavaScript']
projects = ('ChatBot', 'Weather App', 'Portfolio')
unique_numbers = {10, 20, 30, 40}
employee_active = True
print("Employee\n")
print(employee,"\n")
print("Programming Languages\n")
print(programming_language,"\n")
print("Projects\n")
print(projects,"\n")
print("Unique Numbers\n")
print(unique_numbers,"\n")
print("Employee Active\n")
print(employee_active,"\n")
print("-"*30,"\n")
print("Data Types\n")
print("Employee Dictionary : ",type(employee))
print("Programming Languages : ",type(programming_language))
print("Projects : ",type(projects))
print("Unique Numbers : ",type(unique_numbers))
print("Employee Active : ",type(employee_active))
````
