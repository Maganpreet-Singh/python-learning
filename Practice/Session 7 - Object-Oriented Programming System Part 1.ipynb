# Question 1 — Student Class

Create a class named `Student`.

Data Members:
- name
- roll_no
- marks

Methods:
- display_details()
- is_pass()

A student passes if marks are **40 or above**.

### Example Output

```text
Name : Rahul
Roll No : 25
Marks : 81
Status : Pass
```


````python
class Student:
    def __init__(self,name,roll_no,marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks
        self.display_details()
    
    def is_pass(self):
        if self.marks >= 40:
            print(f"Status : Pass")
        else:
            print(f"Status : Fail")

    def display_details(self):
        print(f"Name : {self.name}")
        print(f"Roll No : {self.roll_no}")
        print(f"Marks : {self.marks}")
        self.is_pass()

student = Student("Rahul",25,81)
````

# Question 2 — Bank Account

Create a class named `BankAccount`.

Attributes:
- account_holder
- pin
- balance

Methods:
- deposit(amount)
- withdraw(amount)
- check_balance()

Conditions:
- Do not allow withdrawal if balance is insufficient.
- Print appropriate messages.

### Example

```text
Deposit : 5000

Withdraw : 1200

Current Balance : 3800
```

````python
class BankAccount:

    def __init__(self, name, pin, balance):
        self.account_holder = name
        self.pin = int(pin)
        self.balance = int(balance)

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposit : {amount}")
        else:
            print("Invalid amount.")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdraw : {amount}")
        else:
            print("Insufficient Balance")
   
    def check_balance(self):
        print(f"Current Balance : {self.balance}")


# --- User Interface / Menu Logic ---

def main_menu():
    # Create the object outside the loop
    person = BankAccount("Rahul", 1234, 10000)
    
    while True:
        print("\nWelcome To Bank")
        print("1. Deposit") 
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")
        
        choice = int(input("Enter the option: "))
        
        if choice == 4:
            print("Exiting...")
            break
            
        user_pin = int(input("Enter your Pin: "))
        
        if user_pin == person.pin:
            if choice == 1:
                amount = int(input("Enter the amount to deposit: "))
                person.deposit(amount)
            elif choice == 2:
                amount = int(input("Enter the amount to withdraw: "))
                person.withdraw(amount)
            elif choice == 3:
                person.check_balance()
            else:
                print("Invalid Option")
        else:
            print("Incorrect Pin")

# Run the menu
main_menu()
````

# Question 3 — Rectangle

Create a class named `Rectangle`.

Attributes:
- length
- width

Methods:
- area()
- perimeter()
- display()

### Example Input

```text
Length = 12

Width = 5
```

### Example Output

```text
Area = 60

Perimeter = 34
```

````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        # Return the calculation instead of creating a new variable
        return self.length * self.width

    def perimeter(self):
        # Added the missing '*' for multiplication
        return 2 * (self.width + self.length)
    
    def display(self):
        # Call the methods to get the values
        print(f"Area = {self.area()}")
        print(f"Perimeter = {self.perimeter()}")

# Creating the object and calling display
rect = Rectangle(10, 10)
rect.display()
````

# Question 4 — Employee Salary

Create a class named `Employee`.

Attributes:
- name
- basic_salary

Methods:
- calculate_bonus() → 15% of basic salary
- calculate_net_salary()
- display()

Net Salary = Basic Salary + Bonus

### Example Output

```text
Employee : Aman

Basic Salary : 50000

Bonus : 7500

Net Salary : 57500
```

````python
class Employee:
    def __init__(self, name, basic_salary):
        self.name = name
        self.basic_salary = int(basic_salary)

    def calculate_bonus(self):
        # Calculate and return the bonus as an integer
        return int(0.15 * self.basic_salary)
    
    def calculate_net_salary(self):
        # Calculate using the basic salary and the bonus method
        return self.basic_salary + self.calculate_bonus()
    
    def display(self):
        # Call the methods dynamically to ensure the math is always up-to-date
        print(f"Employee : {self.name}")
        print(f"Basic Salary : {self.basic_salary}")
        print(f"Bonus : {self.calculate_bonus()}")
        print(f"Net Salary : {self.calculate_net_salary()}")

# Creating the object and calling display
employee1 = Employee("Aman", 50000)
employee1.display()
````

# Question 5 — Library Book Management

Create a class named `Book`.

Attributes:
- title
- author
- price
- available (True/False)

Methods:
- issue_book()
- return_book()
- display_details()

Conditions:
- If the book is already issued, display an appropriate message.
- Returning the book should make it available again.

### Example Output

```text
Title : Atomic Habits

Author : James Clear

Price : 599

Status : Available

Book Issued Successfully.

Status : Not Available

Book Returned Successfully.

Status : Available
```

````python
class Book:
    def __init__(self, title, author, price, available=True):
        self.title = title
        self.author = author
        self.price = int(price)
        self.available = available  # True means available, False means issued

    def issue_book(self):
        # Check if the book is currently available
        if self.available:
            self.available = False
            print("Book Issued Successfully.")
        else:
            print("Sorry, this book is currently not available.")

    def return_book(self):
        # If the book is returned, it becomes available again
        if not self.available:
            self.available = True
            print("Book Returned Successfully.")
        else:
            print("This book was not issued out.")

    def display_details(self):
        print(f"Title : {self.title}")
        print(f"Author : {self.author}")
        print(f"Price : {self.price}")
        # Use a quick conditional to print the correct status string
        status = "Available" if self.available else "Not Available"
        print(f"Status : {status}")

# --- Testing the Class ---

# 1. Create the object
book1 = Book("Atomic Habits", "James Clear", 599, True)

# 2. Display initial details
book1.display_details()

# 3. Issue the book
book1.issue_book()
book1.display_details()

# 4. Return the book
book1.return_book()
book1.display_details()
````
